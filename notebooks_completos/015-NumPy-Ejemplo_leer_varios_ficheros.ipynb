Vamos a leer fichero con datos de la contaminación del aire en el Barrio del Pilar en Madrid.

```python
import os
import glob
```

```python
def dada_carpeta_encuentre_pilar(folder_path):
    
    path_to_find = os.path.join(folder_path, "barrio_del_pilar*.csv")
    
    lista_paths_pilar = glob.glob(path_to_find)
    
    return lista_paths_pilar
    
```

```python
# Generar funcion de lectura

def read_datos_barrio_del_pilar(path):
    
    fecha = recibe_path_pilar_devuelve_fecha(path)
    
    data = recibe_path_pilar_devuelve_datos(path)
    
    return fecha, data
```

```python
def recibe_path_pilar_devuelve_fecha(path):
    
    fecha = os.path.basename(path).split("-")[1].split(".")[0]
    
    return fecha
```

```python
def recibe_path_pilar_devuelve_datos(path):
    
    data = np.genfromtxt(path,
                         skip_header=3, delimiter=';',
                         usecols=(2, 3, 4))
    
    return data
```

```python
folder_path = "../data/"

lista_paths_pilar = dada_carpeta_encuentre_pilar(folder_path)

barrio_pilar_dict = dict()

for path in lista_paths_pilar:
    
    print(path)
    
    fecha, data = read_datos_barrio_del_pilar(path)
    
    barrio_pilar_dict[fecha] = data
    
```
